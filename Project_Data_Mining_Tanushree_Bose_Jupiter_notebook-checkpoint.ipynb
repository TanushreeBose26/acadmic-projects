#### 1) Let's do outlier treatment of the column <u>probability_of_full_payment

```python
df_1.probability_of_full_payment.describe() 
```

```python
sns.boxplot(x=df1.probability_of_full_payment)
```

```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=np.percentile(col,[25,75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5 * IQR)
    upper_range= Q3+(1.5 * IQR)
    return lower_range, upper_range
```

```python
lr,ur=remove_outlier(df_1['probability_of_full_payment'])
print("lower range",lr, "and upper range", ur)
```

```python
df_1['probability_of_full_payment']=np.where(df_1['probability_of_full_payment']>ur,ur,df_1['probability_of_full_payment'])
df_1['probability_of_full_payment']=np.where(df_1['probability_of_full_payment']<lr,lr,df_1['probability_of_full_payment'])
```

#### Silhouette Scoring

```python
x_full=data_scaled.loc[:].values
from sklearn.metrics import silhouette_samples, silhouette_score
```

```python
ss={1:0}
for i in range(3,9):
    clusterer=KMeans(n_clusters=i,init='k-means++',random_state=42)
    y=clusterer.fit_predict(x_full) # The higher (up to 1) the better
    s=silhouette_score(x_full,y)
    ss[i]=round(s,5)
    print("The Average Silhouette Score for {}clusters is {}".format(i,round(s,5)))
```

```python
from sklearn.decomposition import PCA
pca_2 = PCA(2) 
plot_columns = pca_2.fit_transform(x_full)
plt.figure(figsize=(12,7))
sns.scatterplot(x=plot_columns[:,1],y=plot_columns[:,0],hue=KMeans(n_clusters=3,
                                                                   random_state=0).fit(x_full).labels_,
                palette='Dark2_r',legend=False)
plt.title('Cluster Plot for 3 Clusters')
plt.show()
```
